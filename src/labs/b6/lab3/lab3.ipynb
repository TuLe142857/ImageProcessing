```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('circles.png', cv2.IMREAD_GRAYSCALE)
SE = np.ones((96, 96), np.uint8)

eroded_img = cv2.erode(img, SE, iterations=1)

labels, label_img = cv2.connectedComponents(eroded_img)
print(f"count = {labels-1}")

plt.figure("lab 4", (15, 10))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("original")

plt.subplot(2, 2, 2)
plt.imshow(eroded_img, cmap='gray')
plt.title("Erosion")

plt.subplot(2, 2, 3)
plt.imshow(label_img, cmap='nipy_spectral')
plt.title(f"Count = {labels-1}")

plt.tight_layout()
```
