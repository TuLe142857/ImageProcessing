```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('bacteria.png', cv2.IMREAD_GRAYSCALE)

SE = np.ones((20, 20), np.uint8)

eroded_img = cv2.erode(img, SE, iterations=1)

plt.figure("lab 1 - Erosion", (15, 10))

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("original")


plt.subplot(1, 2, 2)
plt.imshow(eroded_img, cmap='gray')
plt.title("Erosion")

plt.tight_layout()
```
